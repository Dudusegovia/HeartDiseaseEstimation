```python
import numpy as np
import pandas as pd
```

```python
archive = pd.read_csv('heart.csv')
archive
```

### ChestPainType - Dor no peito 
###### ATA - Angina típica
###### NAP - Anigma atípica
###### ASY - Assintomático
###### TA - Dor toráxica nao específica

### Resting BP
###### Pressão arterial em repouso

### Fasting BS 
###### Glicemia em jejum - 0 -> Açucar < 120 mg/dL

### RestingECG 
###### Resultados eletrocardiograma em repouso
###### Normal - ST: Anormalidade na onda ST-T - LVH: hipertrofia ventricular esquerda

### MaxHR 
###### Frequência cardíaca máxima

### ExerciseAngina
###### Angina induzida por exercício

### oldpeak
###### Depressão do segmento ST (milivolts)

### ST_Slope
###### Inclinação do segmento ST durante o esforço
###### Up - Normal _ Flt - plano (pode indicar isquemia) - Down(geralmente se associa a isquemia sinificativa)

### HeartDisease
###### 0 - sem doença
###### - presença de doença cardíaca


```python
df = archive
df
```

# Tratando dados

```python
print(df['Sex'].unique())
print(df['ChestPainType'].unique())
print(df['RestingECG'].unique())
print(df['ExerciseAngina'].unique())
print(df['ST_Slope'].unique())
```

```python
df['Sex'] = df['Sex'].replace({'M':1, 'F':2})
df['ChestPainType'] = df['ChestPainType'].replace({'ATA': 1, 'NAP': 2, 'ASY': 3, 'TA': 4})
df['RestingECG'] = df['RestingECG'].replace({'Normal':1, 'ST':2, 'LVH':3})
df['ExerciseAngina'] = df['ExerciseAngina'].replace({'N': 1, 'Y': 2})
df['ST_Slope'] = df['ST_Slope'].replace({'Up': 1, 'Flat': 2, 'Down': 3})
df
```

# Inserindo bias

```python
df.insert(0, 'bias', 1)
```

```python
df.to_csv('dados_arrumados.csv', index=False)
arquivo_novo = pd.read_csv('dados_arrumados.csv')
arquivo_novo
```

# Filtrando dados corrompidos

```python
filtro = (arquivo_novo['Cholesterol'] < 30) | (arquivo_novo['MaxHR'] < 30) | (arquivo_novo['RestingBP'] < 30)
df_semdados = arquivo_novo[~filtro]
df_semdados
```

# Definindo funções estatísticas

```python
import copy as cp
def acha_coef(df):
    global coef
    y = df['HeartDisease']
    x = df.drop(columns='HeartDisease')
    xt = x.T
    xtx=xt.dot(x)
    
    inv=np.linalg.pinv(xtx)
    invdot=(inv).dot(xt)
    coef = (invdot).dot(y)
    return coef
def estima_heartdisease(row):
    global coef
    return coef[0] + row["Age"]*coef[1] + row["Sex"]*coef[2] + row["ChestPainType"]*coef[3] + row["RestingBP"]*coef[4] + row["Cholesterol"]*coef[5] + row["FastingBS"]*coef[6] + row["RestingECG"]*coef[7] + row["MaxHR"]*coef[8] + row["ExerciseAngina"]*coef[9] + row["Oldpeak"]*coef[10] + row["ST_Slope"]*coef[11]

```

# Separando 100 lições para estimar

```python
df_para_estimar = df_semdados[0:100]
df_para_estimar
```

# Eliminando as 100 lições selecionadas do banco de dados

```python
df_semdados = df_semdados.drop([i for i in range(100)])
df_semdados
```

# Sorteando 90 lições do banco de dados

```python
df_sorteio=df_semdados.sample(90)
df_sorteio
```

# Achando coeficientes

```python
acha_coef(df_sorteio)
```

# Estimando HeartDisease

```python
estima=round(estima_heartdisease(df_para_estimar))
```

# Comparando o resultado estimado com o resultado real

```python
count=0
for i in df_para_estimar['HeartDisease'] == estima:
    if i:
        count+=1
```

# 92% de acerto

```python
count
```

# Médias de acertos com 100 lições sendo sortedas 100 vezes 

```python
media=0
for i in range(100):
    df_sorteio=df_semdados.sample(100)
    acha_coef(df_sorteio)
    estima=round(estima_heartdisease(df_para_estimar))
    count=0
    for i in df_para_estimar['HeartDisease'] == estima:
        if i:
            count+=1
    media+=count
media/100
```

# 86% de acerto, em média, o modelo proporciona. Com um banco de dados de 640 lições.
